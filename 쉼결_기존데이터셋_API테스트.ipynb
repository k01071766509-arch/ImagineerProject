```python
!pip install -q hume

import time
import pandas as pd
from google.colab import files

from hume.client import HumeClient
from hume.expression_measurement.batch.types import (
    InferenceBaseRequest,
    Models,
    Prosody
)

HUME_API_KEY = ""
client = HumeClient(api_key=HUME_API_KEY)

print("기쁨 음성 파일 1개 업로드")
uploaded = files.upload()
file_path = list(uploaded.keys())[0]

config = InferenceBaseRequest(
    models=Models(
        prosody=Prosody()
    )
)

with open(file_path, "rb") as audio_file:
    job = client.expression_measurement.batch.start_inference_job_from_local_file(
        json=config,
        file=[audio_file]
    )

job_id = job if isinstance(job, str) else job.job_id
print("Job ID:", job_id)

while True:
    details = client.expression_measurement.batch.get_job_details(id=job_id)
    status = details.state.status
    print("현재 상태:", status)

    if status == "COMPLETED":
        break
    if status == "FAILED":
        raise Exception("Hume 분석 실패")

    time.sleep(5)

results = client.expression_measurement.batch.get_job_predictions(id=job_id)

prediction = results[0].results.predictions[0]

emotions = (
    prediction.models.prosody.grouped_predictions[0]
    .predictions[0]
    .emotions
)

top5 = sorted(emotions, key=lambda x: x.score, reverse=True)[:5]

df = pd.DataFrame([
    {
        "파일명": file_path,
        "기존 라벨": "기쁨",
        "Hume 감정": emotion.name,
        "점수": round(emotion.score, 4)
    }
    for emotion in top5
])

display(df)
```

# **Happiness (기쁨)**

```python
!pip install -q hume pandas

import time
import shutil
import pandas as pd
from google.colab import files

from hume.client import HumeClient
from hume.expression_measurement.batch.types import (
    InferenceBaseRequest,
    Models,
    Prosody
)

# ==========================================
# API KEY
# ==========================================
HUME_API_KEY = ""

client = HumeClient(api_key=HUME_API_KEY)

# ==========================================
# 기쁨 음성 3개 업로드
# ==========================================
file_paths = []

for i in range(3):
    print(f"happiness 음성 {i+1}번째 파일 업로드")

    uploaded = files.upload()
    file_path = list(uploaded.keys())[0]
    file_paths.append(file_path)

print("\n원본 파일명")
print(file_paths)

# ==========================================
# Hume 전송용 파일명 변경
# ==========================================
safe_file_paths = []

for i, path in enumerate(file_paths, start=1):
    new_name = f"happiness_{i}.wav"
    shutil.copy(path, new_name)
    safe_file_paths.append(new_name)

print("\nHume 전송 파일명")
print(safe_file_paths)

# ==========================================
# Hume v3 Prosody 설정
# ==========================================
config = InferenceBaseRequest(
    models=Models(
        prosody=Prosody()
    )
)

# ==========================================
# 분석 요청
# ==========================================
audio_files = [open(path, "rb") for path in safe_file_paths]

job = client.expression_measurement.batch.start_inference_job_from_local_file(
    json=config,
    file=audio_files
)

for f in audio_files:
    f.close()

# ==========================================
# Job ID
# ==========================================
if isinstance(job, str):
    job_id = job
elif hasattr(job, "job_id"):
    job_id = job.job_id
elif hasattr(job, "id"):
    job_id = job.id
else:
    job_id = str(job)

print("\nJob ID:", job_id)

# ==========================================
# 완료 대기
# ==========================================
while True:
    details = client.expression_measurement.batch.get_job_details(
        id=job_id
    )

    status = details.state.status
    print("현재 상태:", status)

    if status == "COMPLETED":
        break

    if status == "FAILED":
        raise Exception("분석 실패")

    time.sleep(5)

# ==========================================
# 결과 가져오기
# ==========================================
results = client.expression_measurement.batch.get_job_predictions(
    id=job_id
)

# ==========================================
# 결과 정리
# ==========================================
rows = []

for idx, result in enumerate(results):
    original_file = file_paths[idx]

    if not result.results.predictions:
        rows.append({
            "원본 파일명": original_file,
            "실제 라벨": "happiness",
            "Hume 세부감정": "분석 결과 없음",
            "점수": ""
        })
        continue

    prediction = result.results.predictions[0]

    emotions = (
        prediction.models.prosody.grouped_predictions[0]
        .predictions[0]
        .emotions
    )

    top_emotion = max(
        emotions,
        key=lambda x: x.score
    )

    rows.append({
        "원본 파일명": original_file,
        "실제 라벨": "happiness",
        "Hume 세부감정": top_emotion.name,
        "점수": round(top_emotion.score, 4)
    })

df = pd.DataFrame(rows)

print("\n===== happiness 음성 3개 분석 결과 =====")

display(df)
```

# **감정 결과 분석 정리**



Hume AI는 하나의 음성에 대해 여러 감정 점수를 제공하며, 본 테스트에서는 가장 높은 점수를 기록한 감정을 기준으로 결과를 정리하였다.

기존 감정 데이터셋에서 선별한 감정 샘플을 활용하여 Hume AI Prosody 모델을 통해 총 7개 감정에 대한 음성 감정 분석을 진행하였다.
분석 결과 Hume AI는 직접적인 감정보다는 Sympathy, Distress, Calmness, Tiredness, Contemplation 등의 세부 감정을 주로 반환하였다. 특히 놀람(Surprise)은 Interest, Excitement, Surprise(positive) 등으로 분석되어 실제 감정과 가장 유사한 결과를 보였으며, 혐오(Disgust)와 두려움(Fear) 또한 Distress와 같은 부정 감정 계열로 비교적 적절하게 분석되었다.

반면 분노(Anger)는 Calmness, Contemplation, Confusion 등으로 분석되어 실제 감정 라벨과 차이가 나타났으며, 기쁨(Happiness) 역시 Joy보다는 Concentration, Interest, Calmness 등의 감정으로 분석되는 경향을 보였다. 또한 일부 음성 파일은 발화 길이가 짧아 감정 분석 결과가 추출되지 않는 경우도 확인되었다.

전체적으로 Hume AI는 기존 감정 라벨과 일부 차이를 보였으나, 음성의 톤과 억양 등 비언어적 특징을 보다 세밀하게 분석하는 경향을 확인할 수 있었다.
